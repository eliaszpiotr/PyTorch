## Saving our Data 

To save our data we can use:
* `df.to_csv` - to save data in `.csv` or `.txt` (*in case of saving to `.txt` we can use `sep=''` for better data representation*).
* `df.to_excel` - to save data into Excel

**Note:** `index=False` - skips indexes during saving data



```python
# Saving data into .csv file
df.to_csv('modified.csv', index=False)

# Saving data into Excel file 
df.to_excel('modified.xlsx', index=False)

# Saving data into .txt file 
df.to_csv('modified.txt', index=False, sep='\t')
```

## Filtering Data
Meaning of signs: 
* `&` - and
* `|` - or

`df.reset_index(drop=True, inplace=True)` is a method in   used to reset the index of a DataFrame. 

```python
# df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]

new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] >= 70)]

# new_df.to_csv('filtred.csv', index=False)

new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
filtred_df = df.loc[~df['Name'].str.contains('Mega')] # not containing word 'Mega' in name by using ~
filtred_df.head(10)
```

```python
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I ,regex=True)] # 'frlags=re.I' -> ignores capital letters
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I ,regex=True)]
```

## Aggregate Statistics (Groupby)

Operations we can perform on a dataframe for eg:
* `sum()`
* `mean()`

```python
# Create a new dataframe
df = pd.read_csv('/content/modified.csv')
df
```

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
```

```python
# Get a sum of values from grouped by `Type 1` dataframe 
df.groupby(['Type 1']).sum()
```

```python
# Adding a column with only ones, to make couting easier
df['count'] = 1

# Counting values by column 'count' from gruped by 'Type1' dataframe 
df.groupby(['Type 1']).count()['count']
```

```python
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```

## Working with large amounts of data

Working with largem aamounts of data can be problematic. In that case we can load our data in chuncks to make it easier.

```python
# Create new dataframe
new_df = pd.DataFrame(columns=df.columns)

# Loading data in chuncks
for df in pd.read_csv('/content/modified.csv', chunksize=5):
  # 
  results = df.groupby(['Type 1']).count()

  new_df = pd.concat([new_df, results])

new_df
```
